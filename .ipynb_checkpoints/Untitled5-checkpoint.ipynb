```python
import pandas as pd
import datetime
import time
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
nflx = pd.read_csv("nflx_trades.csv")
```

### Show first 25 entries of 21-day moving average

```python
nflx.price.rolling(window=21).mean().head(25)
```

### Plot graph of last 200 trades

```python
plt.figure()
f, axes = plt.subplots(1, 1)

plt.style.use('ggplot')
plt.plot(nflx.price.tail(200), label="Close")
plt.plot(nflx.price.tail(200).rolling(9).mean(), label="MA 9 days")
plt.plot(nflx.price.tail(200).rolling(21).mean(), label="MA 21 days")
plt.legend(loc="best")
plt.title("NFLX close and MA(9) and MA(21)")
```

### Create a price_prev column, containing the previous price

```python
nflx["price_prev"] = nflx["price"].shift(1)
nflx["ma12"] = nflx["price"].rolling(12).mean()
nflx["ma26"] = nflx["price"].rolling(26).mean()
nflx["macd"] = nflx.ma12 - nflx.ma26
nflx["macd_signal"] = nflx.macd.rolling(9).mean()
```

### Preview a sample of the data

```python
nflx[95:100]
```

### Plotting is complicated, but plot.subplots() divides the figure into 2 rows of 1 plot each

```python
def get_dt_from_ts(ts):
    try:
        value = datetime.datetime.fromtimestamp(ts / 1000)
        exct_time = value.strftime('%Y-%m-%d %H:%M:%S')
        return exct_time
    except:
        return ""
```

```python
nflx["long_dt"] = nflx["ts"].apply(get_dt_from_ts)
```

```python
nflx.tail()
```

```python
plt.figure()
f, axes = plt.subplots(2, 1)
axes[0].plot(nflx.price.tail(200), label = "Close")
axes[1].plot(nflx.macd.tail(200), label="MACD")
axes[1].plot(nflx.macd_signal.tail(200), label="MACD Signal")
axes[0].set_xticklabels(get_dt_from_ts(nflx.ts))
```
