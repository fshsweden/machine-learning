### Read in the Olympic Data and skip the first 4 rows

```python
oo = pd.read_csv("olympics.csv", skiprows=4)
```

### Pick out the 1896 data

```python
f1986 = oo[oo.Edition == 1896]
f2008 = oo[oo.Edition == 2008]
```

## Plot a simple line

```python
plt.figure()
f1986.Sport.value_counts().plot(kind="bar");
plt.legend(loc='best')
```

```python
f2008.Sport.value_counts().plot(kind="bar");
```
